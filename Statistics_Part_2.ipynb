**Q 1 What is hypothesis testing in statistics?**
  - Hypothesis method compares two opposite statements about a population and uses sample data to decide which one is more likely to be correct.

**Q 2 What is the null hypothesis, and how does it differ from the alternative hypothesis?**
  - The null hypothesis (H0) proposes that there is no significant effect or relationship between variables being studied
  - Null hypothesis is differ from alternative hypothesis as the alternative hypothesis (H1 or Ha) suggests that a significant effect or relationship does exist

**Q 3 What is the significance level in hypothesis testing, and why is it important?**
  - In hypothesis testing, the significance level (alpha, denoted by α) is the probability of incorrectly rejecting the null hypothesis when it is actually true
  - Importance of significance level :
    - Controlling for Type I Errors:
       The significance level helps control the rate of Type I errors, which occur when we reject the null hypothesis when it's actually true (false positive).
    - Establishing a Threshold for Significance:
       It defines the level of evidence needed to conclude that a result is statistically significant, i.e., unlikely to be due to random chance.
    - Guiding Decision Making:
      The significance level, along with the p-value, helps researchers decide whether to reject or fail to reject the null hypothesis, ultimately guiding their conclusions about the research question.
    - Standardized Approach:
      Using a pre-determined significance level ensures a consistent standard for evaluating the evidence and making decisions across different experiments.
    - Important Considerations:
      The choice of significance level depends on the context of the research, the potential consequences of Type I and Type II errors, and the specific research question being investigated.

**Q 4 What does a P-value represent in hypothesis testing?**
  - In hypothesis testing, a P-value represents the probability of observing a result as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true

**Q 5 How do you interpret the P-value in hypothesis testing?**
  - In hypothesis testing, the p-value represents the probability of observing data as extreme as, or more extreme than, the data actually obtained, assuming the null hypothesis is true.

**Q 6 What are Type 1 and Type 2 errors in hypothesis testing?**
  - Type I Error (False Positive):This error happens when you reject the null hypothesis (H₀) when it is actually true. In other words, you conclude there's an effect or relationship when there isn't one.
  -Type II Error (False Negative):This error occurs when you fail to reject the null hypothesis when it is actually false. You conclude that there's no effect or relationship when one actually exists.

**Q 7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
  - In hypothesis testing, a one-tailed test focuses on deviations in a single direction from the null hypothesis, while a two-tailed test examines deviations in both directions. One-tailed tests are used when the researcher has a specific direction in mind for the effect (e.g., a new drug will lower blood pressure), whereas two-tailed tests are used when the researcher is simply looking for any difference, regardless of direction (e.g., a new drug will have some effect on blood pressure, either higher or lower).

**Q 8 What is the Z-test, and when is it used in hypothesis testing?**
  - A Z-test is a statistical test used in hypothesis testing to determine if there's a significant difference between a sample mean and a known population mean, or between the means of two groups, when the population variance is known and the sample size is large.
  - Used:-
    - Large sample sizes: Z-tests are generally used when the sample size is 30 or more (n ≥ 30).
    - Known population standard deviation: The Z-test requires that the population standard deviation (σ) is known.
    - Normal distribution: It assumes the population data follows a normal distribution.
    - Comparing a sample mean to a population mean: A Z-test can be used to determine if a sample's mean differs significantly from a known population mean.
    - Comparing the means of two groups: It can also be used to see if there's a significant difference between the means of two separate groups.

**Q 10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?**

```python
def clt_simulation(distribution, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(distribution, size=sample_size)
        sample_means.append(np.mean(sample))
    return sample_means

lambda_param = 2
exponential_dist = np.random.exponential(scale=1/lambda_param, size=10000)


sample_sizes = [5, 10, 30, 100]
num_samples = 1000

plt.figure(figsize=(12, 8))

for i, sample_size in enumerate(sample_sizes):
    sample_means = clt_simulation(exponential_dist, sample_size, num_samples)

    plt.subplot(2, 2, i + 1)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, label=f"n = {sample_size}")  #Added density
    plt.title(f"Distribution of Sample Means (n = {sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.legend()

plt.tight_layout()
plt.show()
```

**Q 14  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution?**

```python
from scipy.stats import norm

def calculate_zscore(data_point, population_mean, population_std_dev):
    z_score = (data_point - population_mean) / population_std_dev
    return z_score

def compare_to_standard_normal(z_score):

    cdf_value = norm.cdf(z_score)

    print(f"The Z-score is {z_score}")
    print(f"The cumulative probability up to this Z-score is: {cdf_value:.4f}")

    if z_score > 0:
        print(f"The data point is {cdf_value * 100:.2f}% above the population mean")
    elif z_score < 0:
        print(f"The data point is {cdf_value * 100:.2f}% below the population mean")
    else:
        print("The data point is equal to the population mean.")


data_point = 75
population_mean = 70
population_std_dev = 5


z_score = calculate_zscore(data_point, population_mean, population_std_dev)
compare_to_standard_normal(z_score)
```

**Q 17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?**

```python
import numpy as np
from scipy.stats import t


np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=50)

confidence_level = 0.95
degrees_freedom = len(data) - 1
sample_mean = np.mean(data)
sample_standard_error = np.std(data, ddof=1) / np.sqrt(len(data))
t_critical = np.abs(t.ppf((1 - confidence_level) / 2, degrees_freedom))
margin_of_error = t_critical * sample_standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level*100:.0f}%): {confidence_interval}")
```

**Q 4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot?**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_data = [72, 74, 70, 76, 73, 71, 75, 78, 77, 79]
population_mean = 70
population_std_dev = 3
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

x = np.linspace(-4, 4, 500)
y = norm.pdf(x, 0, 1)
plt.plot(x, y)

z_critical = norm.ppf(1 - alpha / 2)
x_crit = np.array([-z_critical, z_critical])
y_crit = norm.pdf(x_crit, 0, 1)
plt.plot(x_crit, y_crit, 'ro')
plt.fill_between(x, 0, y, where=(x < -z_critical) | (x > z_critical), color='lightcoral', alpha=0.5)

plt.show()

if p_value <=0.05:
    print("reject the value ")
else:print("fail to reject")
```

**Q 5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(true_mean, sample_size, alpha, num_simulations=1000):
    type1_errors = 0
    for _ in range(num_simulations):
        sample = np.random.normal(loc=true_mean, scale=1, size=sample_size)
        sample_mean = np.mean(sample)
        z_score = (sample_mean - true_mean) / (1 / np.sqrt(sample_size))
        p_value = 2 * (1 - norm.cdf(abs(z_score)))

        if p_value < alpha:
            type1_errors += 1

    type1_error_rate = type1_errors / num_simulations


    true_mean_h1 = true_mean + 0.5
    type2_errors = 0
    for _ in range(num_simulations):
        sample = np.random.normal(loc=true_mean_h1, scale=1, size=sample_size)
        sample_mean = np.mean(sample)
        z_score = (sample_mean - true_mean) / (1 / np.sqrt(sample_size))
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
        if p_value >= alpha:
            type2_errors += 1

    type2_error_rate = type2_errors/ num_simulations


    plt.figure(figsize=(10, 6))
    x = np.linspace(true_mean - 2, true_mean + 2, 500)
    y = norm.pdf(x, loc=true_mean, scale=1 / np.sqrt(sample_size))  # H0 distribution

    plt.plot(x, y, label='Null Hypothesis (H0)')
    plt.axvline(true_mean + norm.ppf(1- alpha/2) / np.sqrt(sample_size), color='red', linestyle='--', label='Critical Region')
    plt.axvline(true_mean - norm.ppf(1- alpha/2) / np.sqrt(sample_size), color='red', linestyle='--')

    x_h1 = np.linspace(true_mean_h1 - 2, true_mean_h1 + 2, 500)
    y_h1 = norm.pdf(x_h1, loc=true_mean_h1, scale=1 / np.sqrt(sample_size)) #H1 distribution
    plt.plot(x_h1, y_h1, label='Alternative Hypothesis (H1)')
    plt.fill_between(x_h1, 0, y_h1, where=(x_h1 <= true_mean + norm.ppf(1- alpha/2) / np.sqrt(sample_size)), color='skyblue', alpha=0.5)

plt.title('Type I and Type II Errors')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

visualize_errors(true_mean=0, sample_size=30, alpha=0.05)
```

**Q 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?**

```python
# Prior probability of the coin being biased (before observing any data)
prior_probability = 0.1  # Assume 10% chance of the coin being biased

# (Assume a biased coin has a 0.7 probability of heads)
likelihood = 0.7**8 * 0.3**2  # Using binomial distribution for this

# Probability of observing the evidence:
# Calculating the probability of getting 8 heads in 10 flips, considering both the biased (H) and unbiased cases:
# This would normally require a sum over all possible bias levels if we want more detail.
# Here, we only have the biased or unbiased case.
evidence = (0.1 * likelihood) + (0.9 * 0.5**10) # Probability of 8 heads under both biased and unbiased cases.


posterior_probability = (prior_probability * likelihood) / evidence

print(f"Posterior probability of biased coin: {posterior_probability}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import warnings
warnings.filterwarnings('ignore')



data = sns.load_dataset('tips')
d = data[["sex","smoker"]]

data_table=pd.crosstab(data['sex'],data['smoker'])

observed_value=data_table.values

print(F'observed value : ',observed_value)

stats_test,p,dof,excepted_value=stats.chi2_contingency(observed_value)
print("excepted value : ",excepted_value)

dof=1
from scipy.stats import chi2
# to combine data zip is used

for i in zip(observed_value,excepted_value):
    print(i)

chi_square_test=sum([(0-e)**2/e for o,e in zip(observed_value,excepted_value)])
chi_square_stats=chi_square_test[0]+chi_square_test[1]
print("chi_square_stats :-",chi_square_stats)

alpha=0.05

chi_square_critical=chi2.ppf(1-alpha,dof)
print("chi_square_critical :=",chi_square_critical)

if chi_square_stats > chi_square_critical:
        print("reject the null hypothesis")
else:print("fail to reject" )

```
